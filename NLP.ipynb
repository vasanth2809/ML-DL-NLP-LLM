```python
import re
clean_corpus = []

for sentence in sentences: # Iterate through sentences instead of individual characters of the corpus
  review = re.sub('[^a-zA-Z]', ' ', sentence) # Apply regex to each sentence
  review = review.lower()
  # Optional: You might want to split the review into words and then join them back
  # review = ' '.join(review.split())
  clean_corpus.append(review)
```

```python
## stemming
import nltk
from nltk.corpus import stopwords # Import stopwords
nltk.download('stopwords') # Download the stopwords resource

for i in clean_corpus:
  words = nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(stemmer.stem(word))
```

```python
## lemmatization
import nltk
from nltk.corpus import stopwords # Import stopwords
nltk.download('stopwords') # Download the stopwords resource

for i in clean_corpus:
  words = nltk.word_tokenize(i)
  for word in words:
    if word not in set(stopwords.words('english')):
      print(lemmatizer.lemmatize(word))
```
